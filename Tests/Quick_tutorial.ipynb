```python
from SMC.smc import plot_coverage
```

Input is:

plot_coverage(DATA, Skymap, plot = True) 
    
##### Make plot=False if you don't want the plot

DATA can either be:

- An array of fits files


```python
plot_coverage(['1.fits','2.fits'], 'S190814bv.fits')
```

- or taking the fits files from a chosen directory

```python
plot_coverage('Directory', 'S190814bv.fits')
```

```python
plot_coverage('Directory', 'S190814bv.fits', plot=False)
```
